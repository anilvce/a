finding the maximum and minimum for the function f(x)

Step 1: Have your function, f(x), and find the derivative, f(x).

Step 2: Set the derivative, f(x), equal to zero.

Step 3: Solve for x  the result(s) are your critical point(s).

Step 4: Find the second derivative, i.e., find f(x). If f(x) < 0 then its a maximum. If f(x) > 0 then its a minimum


```python
#find the maximium and minimum
from sympy import *
x=Symbol('x')

f=x**2
df=diff(f,x)
d2f=diff(df,x)
df_sol=solveset(Eq(df,0), x)
print(df_sol)
for i in df_sol:
  d2f_val=d2f.evalf(subs={x:i})
  if(d2f_val<0):
    print(i,"is maximum")
  if(d2f_val>0):
    print(i,"is minium")
```
